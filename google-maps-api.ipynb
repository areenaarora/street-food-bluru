#### All the necessary imports

```python
import requests
import pandas as pd
import time
import json
import matplotlib.pyplot as plt
import os
from dotenv import load_dotenv

load_dotenv()
```

### Requesting API data from Google maps

#### For the parameters given, I inserted the latitude and longitude you get for Bangalore, a southern Indian city, from googling for it. For the radius, I chose 2000 (it's in meters, so that would be 20 km — a totally arbitrary choice!)

```python
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
API_KEY = os.getenv("API_KEY")
payload={
    "location": "12.9716, 77.5946", #split the city up
    "radius": "20000",
    "type": "restaurant",
    "key": API_KEY
}

page_1_response = requests.get(url, params=payload)
# print(response.text)

page_2_token = page_1_response.json()['next_page_token']
# print(page_2_token)

payload_two={
    "location": "12.9716, 77.5946",
    "radius": "20000",
    "type": "restaurant",
    "key": API_KEY,
    "pagetoken": page_2_token
}
time.sleep(5)
page_2_response = requests.get(url, params=payload_two)
# print(page_2_response.text)

page_3_token = page_2_response.json()['next_page_token']
# print(page_2_token)

payload_three={
    "location": "12.9716, 77.5946",
    "radius": "20000",
    "type": "restaurant", #try food stall here!
    "key": API_KEY,
    "pagetoken": page_3_token
}
time.sleep(5)
page_3_response = requests.get(url, params=payload_three)
# print(page_3_response.text)
print(page_3_response.json()["status"])
```

#### Note that Google Maps API doesn't give out more than 20 results per query and each search can only yield a maximum of 60 results. So here I've used the next page to token to scrape all 60 results possible and then combined them into a dataframe for visualization.

```python
master_list = []

master_list.extend(page_1_response.json()['results'])
master_list.extend(page_2_response.json()['results'])
master_list.extend(page_3_response.json()['results'])
```
